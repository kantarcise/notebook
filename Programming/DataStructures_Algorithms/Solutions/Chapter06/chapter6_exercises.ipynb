```python
# R-6.6 
# Give a precise and complete definition of the concept of matching for
# grouping symbols in an arithmetic expression. Your definition may be recursive

# Every opening symbol (e.g., parentheses, square brackets, curly braces) must
#  be paired with the correct closing symbol. The matching ensures that the expression is syntactically
#   valid and follows the rules of arithmetic operations.
# 
# Base case: An empty expression has matched grouping symbols, as there are no symbols to match.
# 
# Recursive case: For a non-empty expression, it must satisfy the following conditions:
# a. The expression starts with an opening grouping symbol (e.g., '(', '[', '{').
# b. There exists a corresponding closing grouping symbol (e.g., ')', ']', '}') that matches the opening symbol.
# c. The substring between the opening and closing symbols is a well-formed arithmetic expression, and
#  the process continues recursively to ensure that any nested grouping symbols are also properly matched.
# 
# For example, let's take the expression "(2 + [3 * {5 - 1}])". Here is how the recursive matching works:
# 
# The expression starts with '(', so it's an opening symbol.
# The corresponding closing symbol is ']' (to match the opening '['), and the substring "[3 * {5 - 1}]" is a valid expression.
# Within the substring, it starts with '[', so it's an opening symbol.
# The corresponding closing symbol is '}' (to match the opening '{'), and the substring "{5 - 1}" is a valid expression.
# The substring "{5 - 1}" is a well-formed expression.
# Thus, the complete recursive matching ensures that each opening symbol is matched with the correct
#  closing symbol and that the substrings within each pair of symbols are themselves well-formed
#  expressions, ensuring the validity of the overall arithmetic expression.

# here is the code for it

class Empty(Exception):
    pass

class ArrayStack:
    "LIFO implementation of a stack using Python lists"
    def __init__(self):
        "Make a new stack"
        self._data = []
    
    def __len__(self):
        return len(self._data)

    def is_empty(self):
        return len(self._data) == 0
        
    def push(self, element):
        self._data.append(element)

    def top(self):
        if self.is_empty():
            raise Empty("Stack is empty!")
        return self._data[-1]

    def pop(self):
        """Remove and return last element of the stack"""
        if self.is_empty():
            raise Empty("Stack is empty! No items to be popped")
        return self._data.pop()

def is_matching_symbols(expression):
    stack = ArrayStack()

    # Define the matching symbols dictionary
    matching_symbols = {
        ')': '(',
        ']': '[',
        '}': '{'
    }

    for symbol in expression:
        if symbol in '([{':  # Opening symbol
            stack.push(symbol)
        elif symbol in ')]}':  # Closing symbol
            if stack.is_empty() or stack.top() != matching_symbols[symbol]:
                return False # symbols are not matching.
            stack.pop()

    return stack.is_empty()

# Example usage:
expression1 = "(2 + [3 * {5 - 1}])"
expression2 = "(2 + [3 * {5 - 1}])}"
expression3 = "(2 + [3 * {5 - 1])}"

print(is_matching_symbols(expression1))  # Output: True
print(is_matching_symbols(expression2))  # Output: False
print(is_matching_symbols(expression3))  # Output: False

# The important thing here is that 
# we can use stacks in order to match some sort of characters
# Also a key value can also be used in this comparison
```

```python
# lets write ArrayQueue 

class ArrayQueue:
    """FIFO Queue class using python lists"""

    DEFAULT_CAPACITY = 30

    def __init__(self,):
        self._data = [None] * self.DEFAULT_CAPACITY
        # current size of the queue
        self._size = 0
        # current index of the front of the queue
        self._front = 0

    def __len__(self):
        return self._size

    def is_empty(self):
        return self._size == 0

    def first(self):
        """Return but do not remove the element at the front of the queue"""
        if self.is_empty():
            raise Empty("The queue is empty")
        return self._data[self._front]

    def dequeue(self):
        """Remove and return the element at the front of the queue - FIFO"""
        if self.is_empty():
            raise Empty("No element is in the queue!")

        answer = self._data[self._front]

        # GC yourself
        self._data[self._front] = None
        # Adjust new front, making sure we don't surpass length of the queue
        self._front = (self._front + 1) % len(self._data)
        
        # decrement size
        self._size -= 1
        return answer

    def enqueue(self, e):
        """Add element at the end of the queue"""
        if self._size == len(self._data):
            self._resize(2 * self.DEFAULT_CAPACITY)
        
        # find the index after the last element        
        available = (self._front + self._size) % len(self._data)
        
        # set the element to that place
        self._data[available] = e
    
        # increment size    
        self._size += 1

    def _resize(self, cap):
        # we assume cap >= len(self)
        """Resize to a new list of capacity >= len(self)"""
        # keep track of existing list
        old = self._data
        # allocate list with new capacity
        self._data = [None] * cap
        # only consider existing elements
        first_not_none = self._front
        # intentionally shift indices
        for k in range(self._size):
            self._data[k] = old[first_not_none]
            # use old size as modulus, k is incremented need to increment first_not_none as well
            first_not_none = (1 + first_not_none) % len(old)
        # front has been realigned
        self._front = 0

# R-6.10 Consider what happens if the loop in the ArrayQueue. resize method at
# lines 53–55 of Code Fragment 6.7 had been implemented as:

#     for k in range(self._size):
#         self._data[k] = old[k] # rather than old[walk]

# Give a clear explanation of what could go wrong.

# TLDR: old maybe starting with None. so the transfer would be wrong.

# Long answer
# The self._data array would not be correctly realigned, and the 
# elements would not be transferred to the correct positions in the newly resized array.
#  The loop would simply copy the elements from the old array (up to self._size number of elements)
#   sequentially into the new array, starting from index 0. This approach would not
#    consider the circular nature of the queue.

# Let's illustrate this with an example:

# Assume we have a queue with self._size equal to 4, and the underlying array looks like this:

old = [None, 10, 20, 30, 40] # (where None represents empty slots)

# The self._front is at index 1, and the queue is circular.

# With the incorrect implementation (self._data[k] = old[k]), the loop would execute as follows:

# k = 0: self._data[0] = old[0], so self._data[0] becomes None
# k = 1: self._data[1] = old[1], so self._data[1] becomes 10
# k = 2: self._data[2] = old[2], so self._data[2] becomes 20
# k = 3: self._data[3] = old[3], so self._data[3] becomes 30

# After the loop finishes, the new underlying array will look like this:

#  self._data = [None, 10, 20, 30, None]

# As you can see, the realignment did not happen correctly, and there's an extra None in the
#  new array. This issue occurs because the loop does not take into account the circular nature of
#  the queue and does not reposition the elements based on the original _front index.
# 
# Therefore, using the correct implementation with walk as in the original code ensures that 
# the elements are transferred properly, considering the circular arrangement of the 
# queue and the correct front position.
```

```python
# C-6.20 
# Describe a nonrecursive algorithm for enumerating all permutations of the
# numbers {1, 2, . . . , n} using an explicit stack.

def permutations(n):
    """
    Enumerates all permutations of the numbers {1, 2, ..., n} using a nonrecursive
    algorithm.

    Args:
      n: The number of elements to permute.

    Returns:
      A list of all permutations of the numbers {1, 2, ..., n}.
    """

    stack = [[i] for i in range(n)]  # Initialize the stack with individual elements
    # stack =  [[0], [1], [2]]
    permutations = []  # List to store the final permutations

    while stack:
        current_permutation = stack.pop()  # Pop the current partial permutation from the stack

        if len(current_permutation) == n:
            # A permutation of length n is complete, add it to the list of permutations
            permutations.append(current_permutation)
        else:
            # Continue building the permutation by adding unused elements to the end
            unused_elements = [elem for elem in range(n) if elem not in current_permutation]
            for elem in unused_elements:
                stack.append(current_permutation + [elem])

    return permutations

print(permutations(3))

# this was genius!
```

```python
# C-6.24
#  Describe how to implement the stack ADT using a single queue as an
# instance variable, and only constant additional local memory within the
# method bodies. What is the running time of the push(), pop(), and top()
# methods for your design?

# To implement the stack ADT using a single queue as an
#  instance variable, we can use the following approach:
# 
# For the push() operation, we enqueue the new element at the
#  end of the queue and then rotate the elements of the queue
#  so that the newly added element becomes the front of the queue.
# 
# For the pop() operation, we simply dequeue the front element of the queue.
# 
# For the top() operation, we return the front element of the queue without dequeuing it.
# 
# Here's the Python implementation:

from collections import deque

class StackWithQueue:
    def __init__(self):
        self._queue = deque()

    def push(self, element):
        self._queue.append(element)
        # Rotate the queue so that the new element becomes the front
        for _ in range(len(self._queue) - 1):
            self._queue.append(self._queue.popleft())

    def pop(self):
        if not self.is_empty():
            return self._queue.popleft()

    def top(self):
        if not self.is_empty():
            return self._queue[0]

    def is_empty(self):
        return len(self._queue) == 0

    def __len__(self):
        return len(self._queue)

# The running time of each operation for this implementation is as follows:
# 
# push(): O(n) - The for loop rotates the queue elements, which takes O(n) time, where 
# n is the number of elements in the queue.
# 
# pop(): O(1) - The popleft() method on a deque is an atomic operation with O(1) time 
# complexity.
# 
# top(): O(1) - Accessing the front element of the queue using self._queue[0] is a constant 
# time operation.
# 
# Although the push() operation takes O(n) time due to the rotation, the amortized time 
# complexity for a sequence of n push operations is still O(1) on average per push, as each
#  element is rotated only once and dequeued once.
```

```python
# C-6.26 
# Describe how to implement the double-ended queue ADT using two stacks
# as instance variables. What are the running times of the methods?

# To implement the double-ended queue (deque) ADT using two stacks as instance variables, we
#  can use the following approach:
# 
# Use one stack, let's call it head_stack, for operations related to the front of 
# the deque (left side).
# 
# Use another stack, let's call it tail_stack, for operations related to the back of 
# the deque (right side).

class DequeWithStacks:
    def __init__(self):
        self.head_stack = []
        self.tail_stack = []

    def is_empty(self):
        return not self.head_stack and not self.tail_stack

    def __len__(self):
        return len(self.head_stack) + len(self.tail_stack)

    def add_first(self, element):
        self.head_stack.append(element)

    def add_last(self, element):
        self.tail_stack.append(element)

    def delete_first(self):
        if self.head_stack:
            return self.head_stack.pop()
        elif self.tail_stack:
            while self.tail_stack:
                self.head_stack.append(self.tail_stack.pop())
            return self.head_stack.pop()
        else:
            raise Empty("Deque is empty")

    def delete_last(self):
        if self.tail_stack:
            return self.tail_stack.pop()
        elif self.head_stack:
            while self.head_stack:
                self.tail_stack.append(self.head_stack.pop())
            return self.tail_stack.pop()
        else:
            raise Empty("Deque is empty")

    def first(self):
        if self.head_stack:
            return self.head_stack[-1]
        elif self.tail_stack:
            return self.tail_stack[0]
        else:
            raise Empty("Deque is empty")

    def last(self):
        if self.tail_stack:
            return self.tail_stack[-1]
        elif self.head_stack:
            return self.head_stack[0]
        else:
            raise Empty("Deque is empty")

# The running times of the methods in this implementation are as follows:
# 
# add_first(): O(1) - Appending an element to the head_stack takes constant time.
# 
# add_last(): O(1) - Appending an element to the tail_stack takes constant time.
# 
# delete_first(): O(1) amortized - If head_stack is not empty, the operation takes constant time.
#  If head_stack is empty, the elements are transferred from tail_stack to head_stack, which 
# takes O(n) time for n elements. However, when we consider a sequence of n delete_first
#  operations, each element is transferred at most once from 
# tail_stack to head_stack. Therefore, the amortized time complexity is O(1).
# 
# delete_last(): O(1) amortized - If tail_stack is not empty, the operation takes constant
#  time. If tail_stack is empty, the elements are transferred from head_stack to tail_stack, which
#  takes O(n) time for n elements. However, when we consider a sequence of n delete_last
#  operations, each element is transferred at most once from head_stack to tail_stack. 
# Therefore, the amortized time complexity is O(1).
# 
# first(): O(1) - Accessing the top element of head_stack or the first element of tail_stack 
# takes constant time.
# 
# last(): O(1) - Accessing the top element of tail_stack or the first element of head_stack
#  takes constant time.
# 
# Overall, the implementation achieves O(1) amortized time complexity for all deque operations.
```

```python
# C-6.27 Suppose you have a stack S containing n elements and a queue Q that is
# initially empty. Describe how you can use Q to scan S to see if it contains a
# certain element x, with the additional constraint that your algorithm must
# return the elements back to S in their original order. You may only use S,
# Q, and a constant number of other variables.

# To scan the stack S for a certain element x and return the elements back
#  to S in their original order, we can use the following algorithm:
# 
# Initialize a variable found to False.
# 
# Pop elements from S one by one and enqueue them into Q. Keep track of 
# the popped elements in a temporary stack, let's call it temp_stack.
# 
# While dequeueing elements from Q, check if the element matches the target element x. 
# If it does, set found to True.
# 
# After dequeuing all elements from Q, enqueue them back to S in the same order
#  as they were dequeued.
# 
# Finally, enqueue the elements from temp_stack back to S in their original order.

class Empty(Exception):
    pass

def search_and_restore(S, Q, x):
    temp_stack = []
    found = False

    # Scan S and enqueue elements to Q and store them in temp_stack
    while not S.is_empty():
        element = S.pop()
        Q.enqueue(element)
        temp_stack.append(element)

    # Dequeue elements from Q, check for the target element x
    while not Q.is_empty():
        element = Q.dequeue()
        if element == x:
            found = True
        S.push(element)

    # Enqueue elements back to S in the same order as dequeued
    while temp_stack:
        S.push(temp_stack.pop())

    return found

# This algorithm uses the stack S, the queue Q, and the temporary stack temp_stack to 
# scan S for the target element x and return all elements back to S in their original order. 
# The additional variables used are constant and do not depend on the size of the stack or queue.
#  The time complexity of this algorithm is O(n), where n is the 
# number of elements in the stack S.
```

```python
# C-6.28 Modify the ArrayQueue implementation so that the queue’s capacity is
# limited to maxlen elements, where maxlen is an optional parameter to the
# constructor (that defaults to None). If enqueue is called when the queue
# is at full capacity, throw a Full exception (defined similarly to Empty).

class Full(Exception):
    pass

class ArrayQueue:
    """FIFO Queue class using Python lists"""

    DEFAULT_CAPACITY = 10

    def __init__(self, maxlen=None):
        """Make a new queue with optional maximum capacity."""
        self.maxlen = maxlen
        self._data = [None] * self.DEFAULT_CAPACITY
        # current size of the queue
        self._size = 0
        # current index of the front of the queue
        self._front = 0

    def __len__(self):
        return self._size

    def is_empty(self):
        return self._size == 0

    def first(self):
        """Return but do not remove the element at the front of the queue."""
        if self.is_empty():
            raise Empty("The queue is empty")
        return self._data[self._front]

    def dequeue(self):
        """Remove and return the element at the front of the queue - FIFO."""
        if self.is_empty():
            raise Empty("No element is in the queue!")

        answer = self._data[self._front]

        # GC yourself
        self._data[self._front] = None

        # Adjust new front
        self._front = (self._front + 1) % len(self._data)

        # decrement size
        self._size -= 1
        return answer

    def enqueue(self, e):
        """Add element at the end of the queue."""
        if self.maxlen is not None and self._size == self.maxlen:
            raise Full("Queue is full!")

        if self._size == len(self._data):
            self._resize(2 * len(self._data))

        # find the index after the last element
        avail = (self._front + self._size) % len(self._data)

        # set the element to that place
        self._data[avail] = e

        # increment size
        self._size += 1

    def _resize(self, cap):
        """Resize to a new list of capacity >= len(self)."""
        # keep track of the existing list
        old = self._data
        # allocate list with new capacity
        self._data = [None] * cap
        # only consider existing elements
        walk = self._front
        # intentionally shift indices
        for k in range(self._size):
            self._data[k] = old[walk]
            # use old size as modulus
            walk = (1 + walk) % len(old)
        # front has been realigned
        self._front = 0


# With this modification, the ArrayQueue will raise a Full exception when
#  enqueue is called and the queue is at its maximum capacity (maxlen).
```

```python
# C-6.29 In certain applications of the queue ADT, it is common to repeatedly
# dequeue an element, process it in some way, and then immediately en-
# queue the same element. Modify the ArrayQueue implementation to in-
# clude a rotate( ) method that has semantics identical to the combina-
# tion, Q.enqueue(Q.dequeue( )). However, your implementation should
# be more efficient than making two separate calls (for example, because
# there is no need to modify size).

# To implement the rotate() method efficiently in the ArrayQueue class, we
#  can use the following approach:
# 
# Get the front element of the queue.
# Increment the _front index to wrap around the queue if needed.
# Replace the original front element with the newly rotated front element.
# This approach avoids modifying the size or shifting elements in the queue, which
#  makes it more efficient than making separate calls to enqueue and dequeue. Here's the
#  updated ArrayQueue class with the rotate() method:


class Empty(Exception):
    pass

class ArrayQueue:
    """FIFO Queue class using Python lists"""

    DEFAULT_CAPACITY = 10

    def __init__(self):
        "Make a new queue"
        self._data = [None] * self.DEFAULT_CAPACITY
        # current size of the queue
        self._size = 0
        # current index of the front of the queue
        self._front = 0

    def __len__(self):
        return self._size

    def is_empty(self):
        return self._size == 0

    def first(self):
        """Return but do not remove the element at the front of the queue."""
        if self.is_empty():
            raise Empty("The queue is empty")
        return self._data[self._front]

    def dequeue(self):
        """Remove and return the element at the front of the queue - FIFO."""
        if self.is_empty():
            raise Empty("No element is in the queue!")

        answer = self._data[self._front]

        # GC yourself
        self._data[self._front] = None

        # Adjust new front
        self._front = (self._front + 1) % len(self._data)

        # decrement size
        self._size -= 1
        return answer

    def enqueue(self, e):
        """Add element at the end of the queue."""
        if self._size == len(self._data):
            self._resize(2 * len(self._data))

        # find the index after the last element
        avail = (self._front + self._size) % len(self._data)

        # set the element to that place
        self._data[avail] = e

        # increment size
        self._size += 1

    def _resize(self, cap):
        """Resize to a new list of capacity >= len(self)."""
        # keep track of existing list
        old = self._data
        # allocate list with new capacity
        self._data = [None] * cap
        # only consider existing elements
        walk = self._front
        # intentionally shift indices
        for k in range(self._size):
            self._data[k] = old[walk]
            # use old size as modulus
            walk = (1 + walk) % len(old)
        # front has been realigned
        self._front = 0

    def rotate(self):
        """Performs a rotate operation: Q.enqueue(Q.dequeue())"""
        if self.is_empty():
            raise Empty("The queue is empty")
        
        front_element = self._data[self._front]
        self._front = (self._front + 1) % len(self._data)
        self._data[(self._front + self._size - 1) % len(self._data)] = front_element

# With the rotate() method, you can efficiently perform a rotate operation without modifying
#  the size or shifting elements in the queue. This makes the operation more efficient
#  than making separate enqueue and dequeue calls.
```

```python
# P-6.36 When a share of common stock of some company is sold, the capital
# gain (or, sometimes, loss) is the difference between the share’s selling
# price and the price originally paid to buy it. This rule is easy to under-
# stand for a single share, but if we sell multiple shares of stock bought
# over a long period of time, then we must identify the shares actually be-
# ing sold. A standard accounting principle for identifying which shares of
# a stock were sold in such a case is to use a FIFO protocol—the shares
# sold are the ones that have been held the longest (indeed, this is the de-
# fault method built into several personal finance software packages). For
# example, suppose we buy 100 shares at $20 each on day 1, 20 shares at
# $24 on day 2, 200 shares at $36 on day 3, and then sell 150 shares on day
# 4 at $30 each. Then applying the FIFO protocol means that of the 150
# shares sold, 100 were bought on day 1, 20 were bought on day 2, and 30
# were bought on day 3. The capital gain in this case would therefore be
# 100 ·10 + 20 ·6 + 30 ·(−6), or $940. Write a program that takes as input
# a sequence of transactions of the form “buy x share(s) at y each”
# or “sell x share(s) at y each,” assuming that the transactions oc-
# cur on consecutive days and the values x and y are integers. Given this
# input sequence, the output should be the total capital gain (or loss) for the
# entire sequence, using the FIFO protocol to identify shares.

class StockInteraction():
    def __init__(self, action, shares, price):
        self.action = action
        self.shares = shares
        self.price = price
    
def calculate_all_transactions(transactions):
    stack = []                # Create an empty stack to hold buy transactions
    capital_gain = 0          # Initialize the total capital gain
    
    for transaction in transactions:
        if transaction.action == "buy":
            stack.append(transaction)  # Push the current transaction onto the stack
        elif transaction.action == "sell":
            shares_left = transaction.shares  # Initialize the remaining shares to sell
            # While there are shares to sell and the stack is not empty
            while shares_left > 0 and stack:  
                top_transaction = stack[-1]   # Get the top transaction from the stack
                 # Determine the minimum of remaining shares and top transaction shares
                min_shares = min(shares_left, top_transaction.shares)                 
                # Calculate the capital gain for the current sell transaction
                capital_gain += (transaction.price - top_transaction.price) * min_shares
                shares_left -= min_shares  # Update the remaining shares to sell
                top_transaction.shares -= min_shares  # Update the shares of the top transaction
                if top_transaction.shares == 0:
                    # If all shares of the top transaction are sold, remove it from the stack
                    stack.pop()  
    
    return capital_gain  # Return the total capital gain


stocks_moving_day_1 = StockInteraction("buy", 100, 20)
stocks_moving_day_2 = StockInteraction("buy", 400, 24)
stocks_moving_day_3 = StockInteraction("buy", 200, 36)
stocks_moving_day_4 = StockInteraction("sell", 550, 30)

history = [stocks_moving_day_1, stocks_moving_day_2, stocks_moving_day_3, stocks_moving_day_4]

calculate_all_transactions(history)
```

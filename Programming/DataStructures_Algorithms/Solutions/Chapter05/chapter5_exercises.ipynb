```python
# an example from Python Docs: https://docs.python.org/3/library/sys.html#sys.getsizeof

from __future__ import print_function
from sys import getsizeof, stderr
from itertools import chain
from collections import deque
try:
    from reprlib import repr
except ImportError:
    pass

def total_size(o, handlers={}, verbose=False):
    """ Returns the approximate memory footprint an object and all of its contents.

    Automatically finds the contents of the following builtin containers and
    their subclasses:  tuple, list, deque, dict, set and frozenset.
    To search other containers, add handlers to iterate over their contents:

        handlers = {SomeContainerClass: iter,
                    OtherContainerClass: OtherContainerClass.get_elements}

    """
    dict_handler = lambda d: chain.from_iterable(d.items())
    all_handlers = {tuple: iter,
                    list: iter,
                    deque: iter,
                    dict: dict_handler,
                    set: iter,
                    frozenset: iter,
                   }
    all_handlers.update(handlers)     # user handlers take precedence
    seen = set()                      # track which object id's have already been seen
    default_size = getsizeof(0)       # estimate sizeof object without __sizeof__

    def sizeof(o):
        if id(o) in seen:       # do not double count the same object
            return 0
        seen.add(id(o))
        s = getsizeof(o, default_size)

        if verbose:
            print(s, type(o), repr(o), file=stderr)

        for typ, handler in all_handlers.items():
            if isinstance(o, typ):
                s += sum(map(sizeof, handler(o)))
                break
        return s

    return sizeof(o)


##### Example call #####

print("Size of BIG element list DEEP", total_size(my_list_small_items))
print("Size of BIG element list DEEP", total_size(my_list))


# if __name__ == '__main__':
#     d = dict(a=1, b=2, c=3, d=[4,5,6,7], e='a string of chars')
#     print(total_size(d, verbose=True))


# Honestly
# I like deepsgo the best
```

```python
# A Basic Dynamic Array class with help of ctypes. 

import ctypes                                   # provides low-level arrays
import collections

class DynamicArray:
    """A dynamic array class akin to a simplified Python list."""
    def __init__(self):
        """Create an empty array."""
        self._n = 0                                 # count actual elements
        self._capacity = 1                          # default array capacity
        self._A = self._make_array(self._capacity)  # low-level array

    def __len__ (self):
        """Return number of elements stored in the array."""
        return self._n

    def __getitem__ (self, k): 
        """Return element at index k."""
        if not 0 <= k < self._n:
            raise IndexError("invalid index")
        return self._A[k] # retrieve from array

    def append(self, obj): 
        """Add object to end of the array."""
        if self._n == self._capacity: # not enough room
            self._resize(2 * self._capacity) # so double capacity
        self._A[self._n] = obj
        self._n += 1 

    def _resize(self, c): # nonpublic utitity
        """Resize internal array to capacity c."""
        B = self._make_array(c) # new (bigger) array
        for k in range(self._n): # for each existing value
            B[k] = self._A[k]
        self._A = B # use the bigger array
        self._capacity = c

    def _make_array(self, c): # nonpublic utitity
        """Return new array with capacity c."""
        return (c * ctypes.py_object)( ) # see ctypes documentation
```

```python
# is inserting to the index couple elements before the ending is faster than inserting
#  at the beginning in a python list?
# 
# You can use amortized performance anology while explaining it 
# 
# In Python lists, inserting elements at the end is generally faster than 
# inserting at the beginning. This is because lists are implemented as dynamic arrays, 
# which means that appending elements to the end of the list requires amortized O(1) time complexity on average. 
# 
# When you append an element to a list, Python typically allocates a slightly larger 
# block of memory to accommodate potential future growth. This way, it avoids frequently 
# resizing the underlying array, which can be an expensive operation. The allocated memory 
# block allows for efficient appending at the end because there is no need to shift existing elements.
# 
# On the other hand, inserting elements at the beginning of a list requires shifting all 
# existing elements to make room for the new element. This operation has a time complexity 
# of O(n), where n is the length of the list. As the list grows larger, the time required for
#  inserting at the beginning becomes increasingly slower.
# 
# To illustrate this with an amortized performance analogy, consider a scenario where you have
#  a bag that can expand to accommodate more items. Initially, the bag has some space to hold
#  items. When you want to add more items, it is faster and easier to place them at the end of
#  the bag, as there is available space. However, if you want to insert items at the beginning, you
#  need to shift all existing items to make room, which can be time-consuming.
# 
# Therefore, if you frequently need to perform insertions, it is generally more efficient to
#  append elements to the end of the list and then access them using their index. If you have
#  specific requirements for inserting at the beginning, you might consider using other data 
# structures like linked lists or deques, which provide more efficient operations for such scenarios.
```

```python
# why is python elements are implemented as dynamic arrays? 

# Python implements elements as dynamic arrays in order to provide efficient random 
# access and support operations like indexing, slicing, and resizing. Dynamic arrays
#  offer several advantages that make them suitable for the typical use cases of Python lists:
# 
# 1. Constant-time random access: Dynamic arrays allow constant-time access to elements
#  based on their indices. This means that accessing an element at a specific index is fast, regardless 
# of the size of the array. This property is essential for many algorithms and operations that rely 
# on random access, such as indexing, slicing, and iterating over the elements.
# 
# 2. Efficient memory allocation: Dynamic arrays allocate memory in contiguous blocks, which allows
#  for efficient memory management. Python lists dynamically resize by allocating a slightly larger
#  block of memory when needed. This approach minimizes the frequency of memory reallocation, as
#  resizing the array involves copying the existing elements to a new memory location.
# 
# 3. Low memory overhead: Python's dynamic arrays have a relatively low memory overhead compared 
# to other data structures like linked lists. Dynamic arrays store the elements themselves as well
#  as additional metadata, such as the array size and capacity. This metadata allows for efficient
#  indexing and resizing operations without consuming excessive memory.
# 
# 4. Cache locality: Dynamic arrays exhibit good cache locality, which means that accessing adjacent
#  elements in memory is faster due to the CPU cache behavior. This property improves the performance
#  of iterating over the elements or performing operations on contiguous sections of the array.
# 
# By using dynamic arrays, Python can provide a versatile data structure like the list that offers
#  efficient random access, dynamic resizing, and a balance between memory efficiency and performance. 
# However, it's worth noting that there are other data structures available in Python's standard 
# library, such as linked lists (collections.deque), if specific use cases require different
#  trade-offs in terms of performance characteristics.
```

```python
class CaesarCipher():
    """Class for doing encryption and decryption using CaesarCipher"""
    def __init__(self, shift):
        """Construct CeaselCipher using given integer shift for rotation"""
        encoder = [None] * 26
        decoder = [None] * 26
        for k in range(26):
            encoder[k] = chr((k+shift) % 26 + ord("A"))
            decoder[k] = chr((k-shift) % 26 + ord("A"))
        self._forward = "".join(encoder)
        self._backward = "".join(decoder)

    def _encrypt(self, message):
        """Return string encrypted by CaesarCipher"""
        return self._transform(message, self._forward)

    def _decrypt(self, secret):
        """Return string decrypted by CaesarCipher"""
        return self._transform(secret, self._backward)

    def _transform(self, original, code):
        """Utility to perform transformation based on given code string."""
        msg = list(original)
        for k in range(len(msg)):
            if msg[k].isupper():
                j = ord(msg[k]) - ord("A")
                msg[k] = code[j]
        return "".join(msg)

if __name__ == "__main__":
    cipher = CaesarCipher(shift=3)
    message = "THE EAGLE IS IN PLAY, MEET AT JOES"
    coded = cipher._encrypt(message)
    print(f"Secret is : {coded}")
    answer = cipher._decrypt(coded)
    print(f"Message is : {answer}")
```

```python
# R-5.6 
# Our implementation of insert for the DynamicArray class, as given in
# Code Fragment 5.5, has the following inefficiency. In the case when a re-
# size occurs, the resize operation takes time to copy all the elements from
# an old array to a new array, and then the subsequent loop in the body of
# insert shifts many of those elements. 

# Give an improved implementation
# of the insert method, so that, in the case of a resize, the elements are
# shifted into their final position during that operation, thereby avoiding the
# subsequent shifting

# # A Basic Dynamic Array class with help of ctypes. 
# This was the original implementation

import ctypes                                   # provides low-level arrays

class DynamicArray:
    """A dynamic array class akin to a simplified Python list."""
    def __init__(self):
        """Create an empty array."""
        self._n = 0                                 # count actual elements
        self._capacity = 1                          # default array capacity
        self._A = self._make_array(self._capacity)  # low-level array

    def __len__ (self):
        """Return number of elements stored in the array."""
        return self._n

    def __getitem__ (self, k): 
        """Return element at index k."""
        if not 0 <= k < self._n:
            raise IndexError("invalid index")
        return self._A[k] # retrieve from array

    def append(self, obj): 
        """Add object to end of the array."""
        if self._n == self._capacity: # not enough room
            self._resize(2 * self._capacity) # so double capacity
        self._A[self._n] = obj
        self._n += 1 

    def _resize(self, c): # nonpublic utitity
        """Resize internal array to capacity c."""
        B = self._make_array(c) # new (bigger) array
        # copy all the elements from
        # an old array to a new array
        for k in range(self._n): # for each existing value
            B[k] = self._A[k]
        self._A = B # use the bigger array
        self._capacity = c

    def _make_array(self, c): # nonpublic utitity
        """Return new array with capacity c."""
        return (c * ctypes.py_object)( ) # see ctypes documentation

    def insert(self, k , obj):
        """Insert object obj at index k."""
        if self._n == self._capacity: # not enough room
            self._resize(2* self._capacity)
        
        # Shift all elements to make room for kth index
        # self._n-1 because start from last element
        for i in range(self._n-1, k-1, -1): # work backwards
            self._A[i+1] = self._A[i]

        self._A[k] = obj
        self._n += 1
```

```python
# P-5.33 
# Write a Python program for a matrix class that can add and multiply two-
# dimensional arrays of numbers, assuming the dimensions agree appropriately
#  for the operation

class Matrix():
    """A basic Matrix class that has addition and multiplication"""
    def __init__(self, data):
        self.data = data
        # number of rows and columns
        self.rows = len(data)
        self.cols = len(data[0])

    def __add__(self,other):
        if not isinstance(other, Matrix):
            raise TypeError("Cannot add a Matrix with something different than Matrix")
        if (self.rows != other.rows) or (self.cols != other.cols):
            raise ValueError("The dimensions of the matrices do not match.")
        
        # double list comprehension
        result = [[0 for _ in range(self.cols)] for _ in range(self.rows)]
        for i in range(self.rows):
            for j in range(self.cols):
                result[i][j] = self.data[i][j] + other.data[i][j]
        return Matrix(result)

    def __mul__(self,other):
        if not isinstance(other, Matrix):
            raise TypeError("Cannot multiply a Matrix with something different")
        if self.cols != other.rows:
            raise ValueError("The dimensions of the matrices do not allow multiplication.")

        result = [[0 for _ in range(other.cols)] for _ in range(self.rows)]

        for i in range(self.rows):
            for j in range(other.cols):
                for k in range(self.cols):
                    result[i][j] += self.data[i][k] * other.data[k][j]
        return Matrix(result)
    
    def __repr__(self):
        return f"Matrix({self.data})"


# Example usage:
A = Matrix([[1, 2], [3, 4]])
B = Matrix([[5, 6], [7, 8]])

print("Matrix A:")
print(A)

print("Matrix B:")
print(B)

C = A + B
print("Matrix C = A + B:")
print("C is from:", C)
print(type(C))

D = A * B
print("Matrix D = A * B:")
print("D is a:", D)
print(type(D))
```

```python
# P-5.36
#  Redesign the CaesarCipher class as a subclass of the SubstitutionCipher
# from the previous problem

class SubstitutionCipher:
    """Class for doing encryption and decryption using SubstitutionCipher"""

    def __init__(self, forward_mapping):
        """Construct SubstitutionCipher using the given forward mapping for encryption."""
        if len(forward_mapping) != 26:
            raise ValueError("Forward mapping must contain all 26 uppercase letters.")

        self._forward = forward_mapping
        self._backward = self._create_backward_mapping(forward_mapping)

    def _create_backward_mapping(self, forward_mapping):
        """Utility method to create the backward mapping from the forward mapping."""
        backward_mapping = [None] * 26
        for i in range(26):
            char = chr(i + ord("A"))
            index = forward_mapping.find(char)
            backward_mapping[i] = chr(index + ord("A"))
        return "".join(backward_mapping)

    def _transform(self, original, code):
        """Utility to perform transformation based on given code string."""
        trans_table = str.maketrans('ABCDEFGHIJKLMNOPQRSTUVWXYZ', code)
        return original.translate(trans_table)

    def _encrypt(self, message):
        """Return string encrypted by SubstitutionCipher"""
        return self._transform(message, self._forward)

    def _decrypt(self, secret):
        """Return string decrypted by SubstitutionCipher"""
        return self._transform(secret, self._backward)


class CaesarCipher(SubstitutionCipher):
    """Class for doing encryption and decryption using CaesarCipher"""

    def __init__(self, shift):
        """Construct CaesarCipher using the given integer shift for rotation"""
        forward_mapping = "".join([chr((k + shift) % 26 + ord("A")) for k in range(26)])
        super().__init__(forward_mapping)


if __name__ == "__main__":
    cipher = CaesarCipher(shift=3)

    message = "HELLO WORLD"
    coded = cipher._encrypt(message)
    print(f"Secret is : {coded}")

    answer = cipher._decrypt(coded)
    print(f"Message is : {answer}")


# In this updated version, the CaesarCipher class is defined as a subclass of the 
# SubstitutionCipher class by inheriting from it (class CaesarCipher(SubstitutionCipher)).

# The CaesarCipher class overrides the __init__ method of the parent class to generate
#  the forward mapping specific to the Caesar cipher using the given shift. It then
#  calls the parent class's __init__ method using super().__init__(forward_mapping) to initialize
#  the parent class with the forward mapping.

# The encryption and decryption methods (_encrypt and _decrypt) are inherited from
#  the SubstitutionCipher class, so we don't need to redefine them in the CaesarCipher class.
#  The parent class's methods will be used directly.

# Now, when you run the modified program, it should work as before, producing the same output
#  for encryption and decryption using the Caesar cipher.
```

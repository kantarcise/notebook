```python
# A Basic Dynamic Array class with help of ctypes. 

import ctypes                                   # provides low-level arrays

class DynamicArray:
    """A dynamic array class akin to a simplified Python list."""
    def __init__(self):
        """Create an empty array."""
        self._n = 0                                 # count actual elements
        self._capacity = 1                          # default array capacity
        self._A = self._make_array(self._capacity)  # low-level array

    def __len__ (self):
        """Return number of elements stored in the array."""
        return self._n

    def __getitem__ (self, k): 
        """Return element at index k."""
        if not 0 <= k < self._n:
            raise IndexError("invalid index")
        return self._A[k] # retrieve from array

    def append(self, obj): 
        """Add object to end of the array."""
        if self._n == self._capacity: # not enough room
            self._resize(2 * self._capacity) # so double capacity
        self._A[self._n] = obj
        self._n += 1 

    def _resize(self, c): # nonpublic utitity
        """Resize internal array to capacity c."""
        B = self._make_array(c) # new (bigger) array
        for k in range(self._n): # for each existing value
            B[k] = self._A[k]
        self._A = B # use the bigger array
        self._capacity = c

    def _make_array(self, c): # nonpublic utitity
        """Return new array with capacity c."""
        return (c * ctypes.py_object)( ) # see ctypes documentation
```

```python
# is inserting to the index couple elements before the ending is faster than inserting
#  at the beginning in a python list?
# 
# You can use amortized performance anology while explaining it 
# 
# In Python lists, inserting elements at the end is generally faster than 
# inserting at the beginning. This is because lists are implemented as dynamic arrays, 
# which means that appending elements to the end of the list requires amortized O(1) time complexity on average. 
# 
# When you append an element to a list, Python typically allocates a slightly larger 
# block of memory to accommodate potential future growth. This way, it avoids frequently 
# resizing the underlying array, which can be an expensive operation. The allocated memory 
# block allows for efficient appending at the end because there is no need to shift existing elements.
# 
# On the other hand, inserting elements at the beginning of a list requires shifting all 
# existing elements to make room for the new element. This operation has a time complexity 
# of O(n), where n is the length of the list. As the list grows larger, the time required for
#  inserting at the beginning becomes increasingly slower.
# 
# To illustrate this with an amortized performance analogy, consider a scenario where you have
#  a bag that can expand to accommodate more items. Initially, the bag has some space to hold
#  items. When you want to add more items, it is faster and easier to place them at the end of
#  the bag, as there is available space. However, if you want to insert items at the beginning, you
#  need to shift all existing items to make room, which can be time-consuming.
# 
# Therefore, if you frequently need to perform insertions, it is generally more efficient to
#  append elements to the end of the list and then access them using their index. If you have
#  specific requirements for inserting at the beginning, you might consider using other data 
# structures like linked lists or deques, which provide more efficient operations for such scenarios.
```

```python
# why is python elements are implemented as dynamic arrays? 

# Python implements elements as dynamic arrays in order to provide efficient random 
# access and support operations like indexing, slicing, and resizing. Dynamic arrays
#  offer several advantages that make them suitable for the typical use cases of Python lists:
# 
# 1. Constant-time random access: Dynamic arrays allow constant-time access to elements
#  based on their indices. This means that accessing an element at a specific index is fast, regardless 
# of the size of the array. This property is essential for many algorithms and operations that rely 
# on random access, such as indexing, slicing, and iterating over the elements.
# 
# 2. Efficient memory allocation: Dynamic arrays allocate memory in contiguous blocks, which allows
#  for efficient memory management. Python lists dynamically resize by allocating a slightly larger
#  block of memory when needed. This approach minimizes the frequency of memory reallocation, as
#  resizing the array involves copying the existing elements to a new memory location.
# 
# 3. Low memory overhead: Python's dynamic arrays have a relatively low memory overhead compared 
# to other data structures like linked lists. Dynamic arrays store the elements themselves as well
#  as additional metadata, such as the array size and capacity. This metadata allows for efficient
#  indexing and resizing operations without consuming excessive memory.
# 
# 4. Cache locality: Dynamic arrays exhibit good cache locality, which means that accessing adjacent
#  elements in memory is faster due to the CPU cache behavior. This property improves the performance
#  of iterating over the elements or performing operations on contiguous sections of the array.
# 
# By using dynamic arrays, Python can provide a versatile data structure like the list that offers
#  efficient random access, dynamic resizing, and a balance between memory efficiency and performance. 
# However, it's worth noting that there are other data structures available in Python's standard 
# library, such as linked lists (collections.deque), if specific use cases require different
#  trade-offs in terms of performance characteristics.
```

```python
class CaesarCipher():
    """Class for doing encryption and decryption using CaesarCipher"""
    def __init__(self, shift):
        """Construct CeaselCipher using given integer shift for rotation"""
        encoder = [None] * 26
        decoder = [None] * 26
        for k in range(26):
            encoder[k] = chr((k+shift) % 26 + ord("A"))
            decoder[k] = chr((k-shift) % 26 + ord("A"))
        self._forward = "".join(encoder)
        self._backward = "".join(decoder)

    def _encrypt(self, message):
        """Return string encrypted by CaesarCipher"""
        return self._transform(message, self._forward)

    def _decrypt(self, secret):
        """Return string decrypted by CaesarCipher"""
        return self._transform(secret, self._backward)

    def _transform(self, original, code):
        """Utility to perform transformation based on given code string."""
        msg = list(original)
        for k in range(len(msg)):
            if msg[k].isupper():
                j = ord(msg[k]) - ord("A")
                msg[k] = code[j]
        return "".join(msg)

if __name__ == "__main__":
    cipher = CaesarCipher(shift=3)
    message = "THE EAGLE IS IN PLAY, MEET AT JOES"
    coded = cipher._encrypt(message)
    print(f"Secret is : {coded}")
    answer = cipher._decrypt(coded)
    print(f"Message is : {answer}")
```

```python
# R-5.6 
# Our implementation of insert for the DynamicArray class, as given in
# Code Fragment 5.5, has the following inefficiency. In the case when a re-
# size occurs, the resize operation takes time to copy all the elements from
# an old array to a new array, and then the subsequent loop in the body of
# insert shifts many of those elements. Give an improved implementation
# of the insert method, so that, in the case of a resize, the elements are
# shifted into their final position during that operation, thereby avoiding the
# subsequent shifting


# # A Basic Dynamic Array class with help of ctypes. 
# This was the original implementation


import ctypes                                   # provides low-level arrays

class DynamicArray:
    """A dynamic array class akin to a simplified Python list."""
    def __init__(self):
        """Create an empty array."""
        self._n = 0                                 # count actual elements
        self._capacity = 1                          # default array capacity
        self._A = self._make_array(self._capacity)  # low-level array

    def __len__ (self):
        """Return number of elements stored in the array."""
        return self._n

    def __getitem__ (self, k): 
        """Return element at index k."""
        if not 0 <= k < self._n:
            raise IndexError("invalid index")
        return self._A[k] # retrieve from array

    def append(self, obj): 
        """Add object to end of the array."""
        if self._n == self._capacity: # not enough room
            self._resize(2 * self._capacity) # so double capacity
        self._A[self._n] = obj
        self._n += 1 

    def _resize(self, c): # nonpublic utitity
        """Resize internal array to capacity c."""
        B = self._make_array(c) # new (bigger) array
        # copy all the elements from
        # an old array to a new array
        for k in range(self._n): # for each existing value
            B[k] = self._A[k]
        self._A = B # use the bigger array
        self._capacity = c

    def _make_array(self, c): # nonpublic utitity
        """Return new array with capacity c."""
        return (c * ctypes.py_object)( ) # see ctypes documentation

    def insert(self, k , obj):
        """Insert object obj at index k."""
        if self._n == self._capacity: # not enough room
            self._resize(2* self._capacity)
        
        # Shift all elements to make room for kth index
        # self._n-1 because start from last element
        for i in range(self._n-1, k-1, -1): # work backwards
            self._A[i+1] = self._A[i]

        self._A[k] = obj
        self._n += 1
```

```python
# C-5.14 
# The shuffle method, supported by the random module, takes a Python
# list and rearranges it so that every possible ordering is equally likely.
# Implement your own version of such a function. You may rely on the
# randrange(n) function of the random module, which returns a random
# number between 0 and n −1 inclusive.


from random import randrange

def my_shuffle(seq):
    result = []
    for _ in range(len(seq)-1):
        # pop actually returns the element that it bumped
        result.append(seq.pop(randrange(0, len(seq))))
    return result

my_shuffle([1,2,3,4,5,6,7,8,9])
```

```python
# Before moving on
# lets implement Fisher Yates algorithm

from random import randrange

def fisher_yates_shuffle(seq):
    shuffled_seq = seq.copy()
    n = len(shuffled_seq)
    print(shuffled_seq)
    print("Starting shuffle..")
    # decreasing from last element to first element
    for i in range(n-1, 0 , -1):
        j = randrange(0, i+1)
        shuffled_seq[i], shuffled_seq[j] = shuffled_seq[j], shuffled_seq[i]
        print(shuffled_seq)
    print("Result is :")
    return shuffled_seq 


fisher_yates_shuffle([1,2,3,4,5,6,7,8,9])

# both fisheryates and random.shuffle is O(n)
# random.shuffle utilizes fisher yates anyway
```

```python
# C-5.15
# 
#  Consider an implementation of a dynamic array, but instead of copying
# the elements into an array of double the size (that is, from N to 2N) when 
# its capacity is reached, we copy the elements into an array with  [N/4]
# additional cells, going from capacity N to capacity N +  [N/4]. Prove that
# performing a sequence of n append operations still runs in O(n) time in
# this case.


# To prove that performing a sequence of n append operations on the dynamic array, where the capacity is 
# increased by [N/4] when full, still runs in O(n) time, we will analyze the amortized time complexity of the append operation.

# In the original implementation, the append operation has an amortized time complexity of O(1).
#  This means that, on average, each append operation takes constant time, regardless of the size of the dynamic array.

# With the modification where the capacity is increased by [N/4] when full, we can still
#  achieve an O(1) amortized time complexity for the append operation.

# Here's the reasoning:
# - Initially, the dynamic array has a capacity of N.
# - When the array reaches its capacity (N elements), a resize operation is triggered
#  to increase the capacity to N + [N/4].
# - The resize operation copies the N elements into the new array, taking O(N) time.
# - After the resize operation, the array has a new capacity of N + [N/4], which allows for additional
#  append operations without resizing until it reaches the new capacity again.

# Considering the cost of the resize operation and the subsequent append operations, we can see
#  that the cost of the resize operation is amortized over the N-1 append operations that
#  can be performed before reaching the new capacity.

# Therefore, each append operation takes O(1) amortized time, and a sequence of n append operations will run in O(n) time.

# In conclusion, even with the modification to increase the capacity by [N/4] when full, the
#  amortized time complexity of the append operation remains O(1), allowing
#  for a sequence of n append operations to run in O(n) time.
```

```python
# C-5.18 
# Give a formal proof that any sequence of n append or pop operations on
# an initially empty dynamic array takes O(n) time, if using the strategy
# described in Exercise C-5.16.

# To formally prove that any sequence of n append or pop operations on an initially empty 
# dynamic array takes O(n) time using the strategy described in Exercise C-5.16, we
#  need to analyze the time complexity of each operation.
# 
# Exercise C-5.16 strategy suggests doubling the size of the dynamic array when it becomes full and
#  shrinking it to half its capacity when it becomes less than one-fourth full.
# 
# Let's denote n as the number of operations in the sequence, and m as the current capacity of the dynamic array.
# 
# Append Operation:
# 
# If the dynamic array is not full (m > number of elements), the append operation takes O(1)
#  time to add an element to the end of the array.
# 
# If the dynamic array is full (m = number of elements), we need to resize the array to
#  accommodate the new element. Resizing takes O(m) time as it involves creating a
#  new array, copying elements, and deallocating the old array.
# 
# The resizing operation occurs when the dynamic array is full, and it
#  takes O(m) time, where m is the current capacity.
# 
# Since the resizing operation occurs at most n/2 times (when m = number
#  of elements), the total time spent on resizing is O(m * n/2) = O(mn).
# 
# Pop Operation:
# 
# If the dynamic array is not empty, the pop operation takes O(1) time
#  to remove the last element from the array.
# 
# If the dynamic array becomes less than one-fourth full after the pop
#  operation, we need to resize the array to reduce its capacity.
#  Similar to the append operation, resizing takes O(m) time.
# 
# The resizing operation occurs when the dynamic array becomes less than 
# one-fourth full, and it takes O(m) time, where m is the current capacity.
# 
# Since the resizing operation occurs at most n/4 times (when m = number of
#  elements), the total time spent on resizing is O(m * n/4) = O(mn).
# 
# To analyze the total time complexity, we sum the time spent on
#  the append and pop operations, as well as the resizing operations:
# 
# Total Time Complexity = O(n) (time spent on append operations) + O(mn) (time spent on resizing operations) + O(n) (time spent on pop operations)
# 
# Since the resizing operations are the dominant factor in the time complexity, we can simplify the equation:
# 
# Total Time Complexity = O(mn)
# 
# Considering the doubling and halving strategy in Exercise C-5.16, the dynamic 
# array's capacity grows and shrinks geometrically. Hence, m can be expressed as
#  a power of 2, i.e., m = 2^k, where k is the number of resizing operations.
# 
# Since the initial capacity is 1 (m = 2^0), the final capacity after k resizing operations is m = 2^k.
# 
# Thus, the total time complexity can be simplified as:
# 
# Total Time Complexity = O(mn) = O((2^k)n)
# 
# To prove that the time complexity is O(n), we need to show that (2^k)n is bounded by a constant multiple of n.
# 
# Using the fact that the resizing operations occur when the dynamic array is full or
#  less than one-fourth full, we have the following inequality:
# 
# 1/4 * (2^k) ≤ n ≤ 1 * (2^k)
# 
# Taking logarithms on both sides:
# 
# log(1/4 * (2^k)) ≤ log(n) ≤ log(1 * (2^k))
# k + log(1/4) ≤ log(n) ≤ k
# 
# Since k is an integer representing the number of resizing operations, we can ignore
#  the logarithmic term log(1/4) as it is constant. Hence, we have:
# 
# k ≤ log(n)
# 
# Substituting the value of m = 2^k, we get:
# 
# m ≤ 2^(log(n))
# m ≤ n
# 
# Therefore, the total time complexity can be bounded as:
# 
# Total Time Complexity = O((2^k)n) ≤ O(mn) ≤ O(n^2)
# 
# Since m ≤ n, we can conclude that the time complexity is bounded by a constant multiple of n, resulting in:
# 
# Total Time Complexity = O(n)
# 
# Thus, we have formally proved that any sequence of n append or pop operations on an initially
#  empty dynamic array takes O(n) time using the strategy described in Exercise C-5.16.
```
